```python
import pandas as pd
from sklearn.cluster import KMeans # unsupervised clustering approach
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import rcParams
from umap import UMAP  # For UMAP analysis
```

```python
df1 = pd.read_csv("No_Anemia_50HM_alldata.csv") # Load none anemic data
df1
```

```python
df2 = pd.read_csv("Moderate_Anemia_25HM_alldata.csv") # Load moderate anemic data
df2
```

```python
df3 = pd.read_csv("Severe_Anemia_12.5HM_alldata.csv") # Load severe anemic data
df3
```

```python
df = pd.concat([df1, df2, df3], axis=0, ignore_index=True) # Concatenate data vertically
df
```

```python
df= df.drop(['Cat4', 'Unnamed: 0'], axis=1) # drop columns and keep only Parasite Concetrations and wavenumbers 
df
```

```python
X = df.drop(columns=['Cat5']) # drop and keep separate the target variable (Cat5)

scaler = StandardScaler() # Standardize the features
X_standardized = scaler.fit_transform(X)

pca = PCA(n_components=2) # Apply PCA
X_pca = pca.fit_transform(X_standardized)

df_pca = pd.DataFrame(data=X_pca, columns=['PC1', 'PC2']) # we then create DataFrame with the PCA results

df_pca['Cat5'] = df['Cat5'] # to PCA results dataframe add lebels "Cat5"

plt.figure(figsize=(10, 6)) # Figure size, now we draw PCA results 

for category in df_pca['Cat5'].unique():
    category_data = df_pca[df_pca['Cat5'] == category] # take only data points for the current category
    plt.scatter(category_data['PC2'], category_data['PC1'], label=f'Category {category}', marker='o') # Then Plot the data points with a specific marker style based on 'Cat5' category

plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.show()
```

```python

# I am now including the kmeans for clustering and PCA scores for PC1 and PC2

df['Cat5'] = df['Cat5'].astype(str)

# First lets Map 'Cat5' values to proper percentage labels (Parasitemia)
df['Cat5_mapped'] = df['Cat5'].map({
    '6PRS': '6%',
    '0.1PRS': '0.1%',
    '0.002PRS': '0.002%',
    '0.00003PRS': '0.00003%',
    '0PRS': '0%'
})

sns.set(context="paper",
        style="white",
        font_scale=2,
        rc={"font.family": "sans-serif"})

rcParams['font.sans-serif'] = 'Arial' # Set the font as Arial, this is NatureComm requirements


fig, ax = plt.subplots(figsize=(15, 10)) # Fig Size


n_clusters = 5 # set clusters
n_init_value = 100  # Set the number of initializations or bootstraps
kmeans = KMeans(n_clusters=n_clusters, n_init=n_init_value, random_state=None)  # lets apply K means
df['Cluster'] = kmeans.fit_predict(X_standardized)

# Now, i am apply PCA for visualization
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_standardized)

explained_variance_ratio = pca.explained_variance_ratio_ # Now, I investigate explained variance ratio for each principal component


custom_legend_labels = df['Cat5_mapped'].unique() # legend labels

scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=df['Cluster'], cmap='Spectral') # lets plots and color each cluster

#Legend based on customized legend labels
legend_handles = [plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=scatter.cmap(scatter.norm(i)),
                             markersize=11, label=label) for i, label in enumerate(custom_legend_labels)]
plt.legend(handles=legend_handles, loc='lower right')

# This will help me to display explained variance ratios in the xlabel and ylabel
plt.xlabel(f'PC1: {explained_variance_ratio[0]*100:.2f}%', fontsize=16, weight="bold", color='black')
plt.ylabel(f'PC2: {explained_variance_ratio[1]*100:.2f}%', fontsize=16, weight="bold", color='black')
plt.show()
```

```python
# Lets use UMAP instead of PCA

# This first figure will only be for UMAP without clustering, just to get an idea

umap = UMAP(n_components=2, random_state=42)
X_umap = umap.fit_transform(X_standardized)

scatter_umap = plt.scatter(X_umap[:, 0], X_umap[:, 1], cmap='viridis', alpha=0.8) # Plot the UMAP visualization

legend_labels = df['Cat5_mapped'].unique() # legend labels and their order
legend_handles = [plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=scatter_umap.cmap(scatter_umap.norm(i)),
                             markersize=11, label=label) for i, label in enumerate(legend_labels)]
plt.legend(handles=legend_handles, loc='upper right')

plt.xlabel('UMAP Component 1', fontsize=16, weight="bold", color='black')
plt.ylabel('UMAP Component 2', fontsize=16, weight="bold", color='black')
plt.show()


# This Second figure will cluster the dimensionality reduced data with UMAP
n_clusters = 5
kmeans = KMeans(n_clusters=n_clusters, random_state=None) 
df['Cluster'] = kmeans.fit_predict(X_umap)

scatter_clustered_umap = plt.scatter(X_umap[:, 0], X_umap[:, 1], c=df['Cluster'], cmap='viridis', alpha=0.8) # Plot the clustered UMAP visualization

custom_legend_labels = df['Cat5_mapped'].unique() # legend labels

# Legend based on 'Cat5' values
legend_handles = [plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=scatter_clustered_umap.cmap(scatter_clustered_umap.norm(i)),
                             markersize=11, label=label) for i, label in enumerate(custom_legend_labels)]
plt.legend(handles=legend_handles, loc='lower center')

plt.xlabel('UMAP Component 1', fontsize=16, weight="bold", color='black')
plt.ylabel('UMAP Component 2', fontsize=16, weight="bold", color='black')
plt.show()
```
